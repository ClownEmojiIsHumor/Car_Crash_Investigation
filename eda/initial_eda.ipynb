```python
import os
from pathlib import Path
import math

import pandas as pd
import numpy as np

import seaborn as sns 
```

```python
inputDir = Path("../data")
inputFile = Path("2022ncdb.xls")
inputPath = inputDir / inputFile
print(inputPath)

df = pd.read_excel(inputPath)
```

```python
print(df.shape)
# print(df.head)
print(df.columns)
```

```python
# Hacky sol to get some quick counts
def checkCounts(df):
    # counter over the diff things, but dont need to do a full loop (not very df like)
    cnt = dict()

    for column in df.columns:
        cnt[column] = df[column].value_counts()
    

    return cnt
```

```python
counts = checkCounts(df)

for col, counts in counts.items():
    print(col)
    print(counts, end= "\n\n")

```
